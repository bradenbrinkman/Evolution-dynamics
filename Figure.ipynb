```python
############################### Fig.2F: mean-field data for b5=5 ###############################
#################### If original data is not available, processed data will be loaded ###################
def cal_average_std(sol, time_steps, hoppings):
    """
    Calculate the mean and std of hopping rates from the solution of the coupled mean-field equation

    Args:
        sol (array_like): numerical solution of the coupled mean-field equation from ODE solver
        time_steps (int): number of steps
        hoppings (array_like): all possible hopping rates in the discretized space

    Returns:
        [tuple of ndarrays]: mean hopping rate and its std at each time step
    """
    fitness = [[np.sum(sol[i][:, t]) for i in range(len(hoppings))]
               for t in range(time_steps)]
    prob = [fitness[t]/np.sum(fitness[t]) for t in range(time_steps)]
    avehop = np.array([np.sum([prob[t][i]*np.array(hoppings[i])
                               for i in range(len(hoppings))], 0) for t in range(time_steps)])
    variance = np.array([np.sum([prob[t][i]*np.array(hoppings[i])**2 for i in range(
        len(hoppings))], 0)-avehop[t]**2 for t in range(time_steps)])
    return avehop, np.sqrt(variance)


t_eval = np.linspace(0, 700, 701)
# load the low-density mean-field numerical solution
if LOAD_ORIGINAL_DATA:
    try:
        with open('./evolve_data/mean_field_high_density/density_3', "rb") as fp:
            full_sol = pickle.load(fp)
    except FileNotFoundError:
        print("Original data not availbale, loading processed data...")
        with open('./processed_data/fig2f', "rb") as fp:
            ave_meanfield_standard, std_meanfield_standard = pickle.load(fp)
        print("Fig.2F data loaded successfully from processed data!")
    else:
        x = [0, 2, 4, 6, 8, 10]
        hopping_discrete = np.array(
            [p for p in itertools.product(x, repeat=5)])
        ave_meanfield_standard, std_meanfield_standard = cal_average_std(
            full_sol.reshape(-1, 5, 2001)[:, :, :701], 701, hoppings=hopping_discrete)  # load first 701 iterations
        print("Panel C data loaded successfully from orignial data!")
else:
    print("Original data not availbale, loading processed data...")
    with open('./processed_data/fig2f', "rb") as fp:
        ave_meanfield_standard, std_meanfield_standard = pickle.load(fp)
    print("Fig.2F data loaded successfully from processed data!")
```

```python
def func(x, a, b, c, d):
    ''' Sigmoid function used to fit the transition curve'''
    return a / (1 + np.exp(-c * (x-d))) + b
```

```python
############################### Panel C: mean-field data for b5=5 ###############################
#################### If original data is not available, processed data will be loaded ###################
def cal_average_std(sol, time_steps, hoppings):
    """
    Calculate the mean and std of hopping rates from the solution of the coupled mean-field equation

    Args:
        sol (array_like): numerical solution of the coupled mean-field equation from ODE solver
        time_steps (int): number of steps
        hoppings (array_like): all possible hopping rates in the discretized space

    Returns:
        [tuple of ndarrays]: mean hopping rate and its std at each time step
    """
    fitness = [[np.sum(sol[i][:, t]) for i in range(len(hoppings))]
               for t in range(time_steps)]
    prob = [fitness[t]/np.sum(fitness[t]) for t in range(time_steps)]
    avehop = np.array([np.sum([prob[t][i]*np.array(hoppings[i])
                               for i in range(len(hoppings))], 0) for t in range(time_steps)])
    variance = np.array([np.sum([prob[t][i]*np.array(hoppings[i])**2 for i in range(
        len(hoppings))], 0)-avehop[t]**2 for t in range(time_steps)])
    return avehop, np.sqrt(variance)


t_eval = np.linspace(0, 700, 701)
# load the low-density mean-field numerical solution
if LOAD_ORIGINAL_DATA:
    try:
        with open('./evolve_data/mean_field_high_density/density_3', "rb") as fp:
            full_sol = pickle.load(fp)
    except FileNotFoundError:
        print("Original data not available, loading processed data...")
        with open('./processed_data/fig2f', "rb") as fp:
            ave_meanfield_standard, std_meanfield_standard = pickle.load(fp)
        print("Panel C data loaded successfully from processed data!")
    else:
        x = [0, 2, 4, 6, 8, 10]
        hopping_discrete = np.array(
            [p for p in itertools.product(x, repeat=5)])
        ave_meanfield_standard, std_meanfield_standard = cal_average_std(
            full_sol.reshape(-1, 5, 2001)[:, :, :701], 701, hoppings=hopping_discrete)  # load first 701 iterations
        print("Panel C data loaded successfully from orignial data!")
else:
    print("Original data not available, loading processed data...")
    with open('./processed_data/fig2f', "rb") as fp:
        ave_meanfield_standard, std_meanfield_standard = pickle.load(fp)
    print("Panel C data loaded successfully from processed data!")
```

```python
# functions of calculate spatial covariance from origial data
def get_num(a, n):
    """
    Count agent numbers at all the locations

        Parameters: 
            a (array_like): list of agent object;
            n (int): size of the square environment
        Return:
            num (array_like): entry at i,j represents the number of agent at location (i,j)
    """
    num = np.zeros((n, n), dtype=int)
    for pos in [a[i].__dict__['pos'] for i in range(len(a))]:
        num[pos] += 1
    return num

def spatial_corr(data, resource, birth, n=25, norm=True):
    """
    Calculate the spatial correlation coefficients
    Arguments:
        norm [boolean]: True for normalized correlation coefficient; False for unnormalized covaraince
    """
    size = data.shape[0]
    corr = np.zeros((n, n))
    if birth == -1:
        loc = [pos for pos, b in np.ndenumerate(data)]
    else:
        loc = [pos for pos, b in np.ndenumerate(data) if b == birth]  # and data[pos] == 1
    for i in range(int(-n/2), int(n/2)+1):
        for j in range(int(-n/2), int(n/2)+1):
            if norm:
                corr[i, j] = (np.mean([data[x, y]*data[(x+i) % size, (y+j) % size] for x, y in loc]
                                  )-np.mean([data[x] for x in loc])*np.mean([data[(x+i) % size, (y+j) % size] for x, y in loc]))/np.std([data[x] for x in loc])/np.std([data[(x+i) % size, (y+j) % size] for x, y in loc])
            else:
                corr[i, j] = (np.mean([data[x, y]*data[(x+i) % size, (y+j) % size] for x, y in loc]
                                  )-np.mean([data[x] for x in loc])*np.mean([data[(x+i) % size, (y+j) % size] for x, y in loc]))
    corr = np.roll(corr, int(n/2), axis=1)
    corr = np.roll(corr, int(n/2), axis=0)
    return corr


def corr_heatmap(data, resource, birth, n, ax, norm=True, imshow_norm=None, suppress_center=True):
    """
    Plot the heatmap of the spatial correlation function with given spatial data

    Arguments:
        data [2d numpy array]: the spatial data where correlation function is to be calculated
        resource [2d numpy array]: the resource level information that correlation calculation can be conditioned on 
        birth [int]: if -1, the correlation is calculated for all locations; else, the correlation is calculated for that resource level
        n [int]: odd number, the size of the correlation map to be calculated
        suppress_center [boolean]: if Ture, the center of the correlation matrix is replaced with 0; else unchanged
    """
    corr = spatial_corr(data, resource, birth, n, norm=norm)
    if suppress_center:
        corr[int(n/2), int(n/2)] = 0

    im = ax.imshow(corr, extent=(-int(n/2)-0.5, int(n/2) +
           0.5, -int(n/2)-0.5, int(n/2)+0.5), norm=imshow_norm)
    ax.set_xticks(ticks=[i for i in range(-int(n/2), int(n/2)+1, 4)])
    ax.set_yticks(ticks=[i for i in range(-int(n/2), int(n/2)+1, 4)])
    ax.tick_params(axis='x', labelsize=12)
    ax.tick_params(axis='y', labelsize=12)
    return im


def corr_heatmap_average_over_trials(density, size, ax, imshow_norm=None, norm=False, n=25):
    cov_trials = np.zeros((n,n))
    for trial in range(20):
        original_data_path = f'./evolve_data/high_density/density_{density}'
        with open(original_data_path+"/trial_{}/iteration_{}".format(trial,20000), "rb") as fp:   # Unpickling
            data = pickle.load(fp)
        with open(original_data_path+"/trial_{}/resource.txt".format(trial), "rb") as fp:   # Unpickling
            resource = pickle.load(fp)
        num = np.array(get_num(data, size))
        cov_trials += spatial_corr(num, resource, -1, n=n, norm=norm)

    corr = cov_trials/20
    corr[int(n/2), int(n/2)] = 0
    im = ax.imshow(corr, extent=(-int(n/2)-0.5, int(n/2) +
           0.5, -int(n/2)-0.5, int(n/2)+0.5), norm=imshow_norm)
    ax.set_xticks(ticks=[i for i in range(-int(n/2), int(n/2)+1, 4)])
    ax.set_yticks(ticks=[i for i in range(-int(n/2), int(n/2)+1, 4)])
    ax.tick_params(axis='x', labelsize=12)
    ax.tick_params(axis='y', labelsize=12)
    return im


def cov_average_over_trials(density, size, norm=False, n=25, center_suppressed=True):
    cov_trials = np.zeros((n,n))
    for trial in range(20):
        original_data_path = f'./evolve_data/high_density/density_{density}'
        with open(original_data_path+"/trial_{}/iteration_{}".format(trial,20000), "rb") as fp:   # Unpickling
            data = pickle.load(fp)
        with open(original_data_path+"/trial_{}/resource.txt".format(trial), "rb") as fp:   # Unpickling
            resource = pickle.load(fp)
        num = np.array(get_num(data, size))
        cov_trials += spatial_corr(num, resource, -1, n=n, norm=norm)

    corr = cov_trials/20
    if center_suppressed:
        corr[int(n/2), int(n/2)] = 0
    return corr


def max_variance(data, resource, birth=-1, norm=False):
    n = 3
    size = data.shape[0]
    corr = np.zeros((n, n))
    if birth == -1:
        loc = [pos for pos, b in np.ndenumerate(data)]
    else:
        loc = [pos for pos, b in np.ndenumerate(data) if b == birth]  # and data[pos] == 1
    for i in range(int(-n/2), int(n/2)+1):
        for j in range(int(-n/2), int(n/2)+1):
            if norm:
                corr[i, j] = (np.mean([data[x, y]*data[(x+i) % size, (y+j) % size] for x, y in loc]
                                  )-np.mean([data[x] for x in loc])*np.mean([data[(x+i) % size, (y+j) % size] for x, y in loc]))/np.std([data[x] for x in loc])/np.std([data[(x+i) % size, (y+j) % size] for x, y in loc])
            else:
                corr[i, j] = (np.mean([data[x, y]*data[(x+i) % size, (y+j) % size] for x, y in loc]
                                  )-np.mean([data[x] for x in loc])*np.mean([data[(x+i) % size, (y+j) % size] for x, y in loc]))
    return (corr[1,0] + corr[0,1] + corr[-1,0] + corr[0,-1])/4
```
